# Plotting the dataset

```python
from mpl_toolkits.mplot3d import Axes3D
# Applying PCA to reduce to project the data in 2 dimensions
pca = PCA(n_components=2)
X_scaled = preprocessing.scale(train_X)
projected_data = pca.fit_transform(X_scaled)
colors = ['blue', 'red']
#fig = plt.figure()
#ax = fig.add_subplot(111, projection='3d')
plt.scatter(projected_data[:, 0], projected_data[:, 1], c=labels,
            cmap = matplotlib.colors.ListedColormap(colors),
           alpha = 0.5)
plt.show()
```

# Proportion of positive and negative examples

```python
nb_positives = len(labels[labels == 1])
nb_negatives = len(labels[labels == 0])
nb_samples = len(labels)
print("Number of positive examples:", nb_positives)
print("Number of negative examples:", nb_negatives)
print("Number of training examples:", nb_samples)
proportions = {'Fraud': [nb_positives*100/nb_samples],
               'Non Fraud': [nb_negatives*100/nb_samples]}
proportions = pd.DataFrame(data = proportions, index=["Percentage(%)"])
proportions
```

# Applying SVM on the dataset

```python
from sklearn import svm

clf = svm.LinearSVC()
clf.fit(train_X.values, labels)
clf.predict(train_X.values)
```

```python
# Accuracy
(10000 - len(labels[clf.predict(train_X.values) != labels]))/10000
```
